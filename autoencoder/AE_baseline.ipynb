```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

### What's the MSE if reconstructed input is simply the mean of sample features?

```python
PCA_data = "./data/mRNA_PCA_{0}_variance_StandardScaled.csv"
for v in [0.2, 0.6, 0.7, 0.8, 0.9]:
    df = pd.read_csv(PCA_data.format(v), index_col="sample_id")
    x = df.as_matrix()
    x_recon = np.array([list(df.mean())] * df.shape[0])
    mse = np.mean((x-x_recon)**2)
    print("mse:", mse)
    plt.hist(df.mean(),bins=50)
    plt.show()    
```

```python
df = pd.read_hdf("./data/mRNA_lognorm_StandardScaled.hdf", index_col="sample_id")
x = df.as_matrix()
x_recon = np.array([list(df.mean())] * df.shape[0])
mse = np.mean((x-x_recon)**2)
print("mse:", mse)
plt.hist(df.mean(),bins=50)
plt.show()    
```
