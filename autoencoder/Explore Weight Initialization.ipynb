```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=DeprecationWarning)

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def leaky_relu(x):
    if x >=0:
        return x
    else:
        return 0.2*x
```

```python
n_in = 16000
n_out = n_in
n_hidden = 1000

with tf.Session() as sess:
    w1 = sess.run(tf.truncated_normal(shape=(n_in, n_hidden)))
    w2 = sess.run(tf.truncated_normal(shape=(n_hidden, n_out)))
x = np.random.normal(size=(1, n_in))
b1 = np.ones(shape=(1, n_hidden))*0.1
z = np.matmul(x, w1) + b1
act = np.vectorize(leaky_relu)(z)
b2 = np.ones(shape=(1, n_out))*0.1
y = np.matmul(act, w2) + b2
```

```python
f = plt.figure(1)
ax=plt.hist(x.flatten(), bins=100)
plt.title("X")

f = plt.figure(2)
ax=plt.hist(act.flatten(), bins=100)
plt.title("activation")

f = plt.figure(3)
ax=plt.hist(y.flatten(), bins=100)
mse = mean_squared_error(x, y)
plt.title("y, mse={0:.2f}".format(mse))
```

### try xavier initialization

```python
n_in = 16000
n_out = n_in
n_hidden = 1000

w1 = np.random.normal(size=(n_in, n_hidden), scale=(2/(n_in+n_out))**0.5)
w2 = np.random.normal(size=(n_hidden, n_out), scale=(2/(n_in+n_out))**0.5)
x = np.random.normal(size=(1, n_in))
b1 = np.ones(shape=(1, n_hidden))*0.1
z = np.matmul(x, w1) + b1
act = np.vectorize(leaky_relu)(z)
b2 = np.ones(shape=(1, n_out))*0.1
y = np.matmul(act, w2) + b2
```

```python
f = plt.figure(1)
ax=plt.hist(x.flatten(), bins=100)
plt.title("X")

f = plt.figure(2)
ax=plt.hist(act.flatten(), bins=100)
plt.title("activation")

f = plt.figure(3)
ax=plt.hist(y.flatten(), bins=100)
mse = mean_squared_error(x, y)
plt.title("y, mse={0:.2f}".format(mse))
```
