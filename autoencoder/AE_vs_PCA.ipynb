```python
import pickle
import numpy as np
import pandas as pd

from bokeh.io import output_notebook
from bokeh import palettes
from bokeh.plotting import figure, show
from matplotlib import pyplot as plt

output_notebook()
%matplotlib inline
```

```python
colors = palettes.Category10_10
```

```python
node_sizes = [8000, 4000, 2000, 1000, 500, 250, 100, 50, 20, 10, 5, 2]
node_sizes += [1500, 900, 800, 600, 400, 300, 200, 90, 80, 70, 60, 40, 30]
node_sizes = sorted(node_sizes)
col_index = pd.MultiIndex.from_product([node_sizes, ["train", "val"]])
results = pd.DataFrame(index=range(2000), columns=col_index)
for node_size in node_sizes:
    train_mse = np.load("../results/AE/{0}/train_mse.npy".format(node_size))
    val_mse = np.load("../results/AE/{0}/val_mse.npy".format(node_size))
    results.loc[:, (node_size,"train")] = train_mse
    results.loc[:, (node_size,"val")] = val_mse
```

```python
for node_size in node_sizes:
    p = figure(height=400, width=800, title="num hidden nodes: "+str(node_size))
    p.line(x=results.index, y=results[node_size]["train"], 
             color=colors[0], legend="train mse")
    p.line(x=results.index, y=results[node_size]["val"], 
             color=colors[1], legend="validation mse")
    show(p)
```

```python
best_val_error = results.swaplevel(i=0, j=1, axis=1)["val"].min()
```

```python
p = figure(height=500, width=500, x_axis_type="log")
p.line(x=best_val_error.index, y=list(best_val_error))
p.circle(x=best_val_error.index, y=list(best_val_error))
show(p)
```
