```python
from bokeh import palettes
from bokeh.io import output_notebook
from bokeh.plotting import figure, show

output_notebook()
```

```python
color20 = palettes.Category20_20
color20b = palettes.Category20b_20
color20c = palettes.Category20c_20
greys = palettes.grey(9)
reds = palettes.Reds6
```

```python
p = figure(height=300, width=600)
for i, each_color20 in enumerate([color20, color20b, color20c]):
    for j, color in enumerate(each_color20):
        p.square(j, i, color=color, size=18)
show(p)
```

```python
p = figure(height=150, width=600)
for i, scale in enumerate([greys, reds]):
    for j, color in enumerate(scale):
        p.square(j, i, color=color, size=18)
p.toolbar_location = None
show(p)
```

```python
color32 = (color20c[:4] + color20[2:4] + color20c[8:12] + 
    color20[6:8] + color20c[12:] + color20[10:12] + color20b[-4:] + color20[-4:] + [greys[2], reds[0]])
p = figure(height=100, width=700)
for j, color in enumerate(color32):
    p.square(j, i, color=color, size=16)
p.toolbar_location = None
show(p)
```
